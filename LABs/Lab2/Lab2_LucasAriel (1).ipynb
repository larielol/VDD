```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import files
import io

#Carregamento dos dados(usando o pandas):
turma2024 = pd.read_csv('/content/dataset-turma-2024.csv', delimiter=";")

print(turma2024.head())

plt.figure(figsize=(10, 6))
languages = turma2024['Marque as linguagens de programação que você já teve algum contato prático:'].str.split(',').explode()
languages.value_counts().plot(kind='bar', color='royalblue', edgecolor='black', width=1.0)
plt.xlabel('Linguagens de Programação') #Eixo x
plt.ylabel('Quantidade de Alunos') #Eixo Y
plt.title("Linguagens de Programação Conhecidas")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

    História a ser contada: Dada a hipótese científica de que existe relação entre a quantidade de semestres para se formar, a quantidade de linguagens de programação e/ou a experiência de participação em P&D, a história a ser contada deve confirmar ou refutar essa hipótese a partir da amostra analisada. Caso confirmada, a relação seria forte ou fraca?!

    Público-alvo: alunos da disciplina de visualização de dados;

    Ações esperadas pelo público: apenas entender essa possível relação e adquirir esse conhecimento revelado a partir dos dados;

    Mapeamento dos dados necessários para a visualização: aí é contigo!

    Tipo de visualização: Mapas de calor;

    Esboço da visualização: não é obrigatório. Você pode esboçar no papel ou fazer uma prova de conceito já em Python e ir evoluindo no notebook;

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import files
import io


#Carregamento dos dados(usando o pandas):
turma2024 = pd.read_csv('/content/dataset-turma-2024.csv', delimiter=";")

# Nomes das colunas
turma2024 = turma2024.rename(columns={
    'Quantos semestres faltam para você se formar?': 'semestres_faltando',
    'Marque as linguagens de programação que você já teve algum contato prático:': 'qtd_linguagens',
    'Você já teve experiência de participação em projetos de Pesquisa e Desenvolvimento?': 'experiencia_pesquisa'
})


turma2024['semestres_faltando'] = pd.to_numeric(turma2024['semestres_faltando'], errors='coerce')
turma2024['qtd_linguagens'] = turma2024['qtd_linguagens'].apply(lambda x: len(str(x).split(',')))
turma2024['experiencia_pesquisa'] = turma2024['experiencia_pesquisa'].map({'Sim': 1, 'Não': 0})

# Gráfico 1: Linguagens Conhecidas e Tempo para Formação
plt.figure(figsize=(10, 5))
sns.heatmap(pd.crosstab(turma2024['qtd_linguagens'], turma2024['semestres_faltando']), annot=True, cmap='YlOrRd', fmt='d')
plt.ylabel('Quantidade de Linguagens')
plt.xlabel('Quantidade de Semestres')
plt.title('Linguagens Conhecidas vs. Quanto falta para se formar')
plt.show()

# Gráfico 2: Experiência em P&D e Tempo para Formação
plt.figure(figsize=(10, 5))
sns.heatmap(pd.crosstab(turma2024['experiencia_pesquisa'], turma2024['semestres_faltando']), annot=True, cmap='YlOrRd', fmt='d')
plt.ylabel('Experiência em Projetos')
plt.xlabel('Quantidade de Semestres')
plt.title('Experiência em P&D vs. Quanto falta para se formar')

plt.show()
```
